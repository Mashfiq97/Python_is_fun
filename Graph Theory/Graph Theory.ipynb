### Task 01

Nora has been taken to “HighLand” for the first Level of the competition. Here there  are n warriors with no weapons in different fixed positions. Nora has been given  with a weapon and placed in position 0, other positions are 1, 2, 3,…, n-1. There are many connections among these positions. If there is a connection between position u  and position v, Nora can go to position v from u (or vice versa) with 1 move. She can  kill the warrior after reaching that position. Now the task of Nora is to kill warrior  “Lina” who is in position ‘x’ Your task is to help Nora with the minimum number of  
moves Nora needs to make to go to Lina.




```python
def BFS(dict1, start, finish):
        s=start
        visited_nodes = [] 
        queue = [(s,[s])]
        
        while len(queue)!=0:
            node, edge = queue.pop(0)
            visited_nodes.append(node)
           
            for i in dict1[node]:
                if (i not in visited_nodes) and (i != finish):
                    queue.append((i, list(edge + [i])))
                    visited_nodes.append(i)
                    #print(queue)
                elif (i == finish):
                    #print(list(edge + [finish]))
                    return len(list(edge + [finish]))-1
                else:
                    continue

file = open("file.txt")
#print(file.read())
for i in range(2):
    total_lines = int(file.readline().strip())

nodes_dictionary = {}
for i in range(int(total_lines)):
    values=file.readline().strip().split(" ")
    key, node = int(values[0]), int(values[1])
    if key in nodes_dictionary.keys():
        nodes_dictionary[key].append(node)
    else:
        nodes_dictionary[key] = [node] 
#print(nodes_dictionary) 


Nora = 0                    
Lina = int(6)                        
moves = BFS(nodes_dictionary,Nora, Lina)
print(moves)

```

### Task 02

Nora has now been taken to “HigherLand” for Level 2 of the competition. Here she discovered  that she will be competing with another participant “Lara”. Here also there are n warriors in  different fixed positions (position 0 through position n-1). Nora is now placed in position ‘p’ and  Lara is placed in position ‘q’. Both of them have to kill the warrior Lina in position ‘x’. The one  who goes to position ‘x’ first, wins Level 2. 



For n different positions and m connections (u<->v) among the positions, your task is to  determine who can go to ‘x’ first –Nora or Lara? Where Nora is in position ‘p’ and Lara is in  position ‘q’. If both win, print both.

```python
def BFS(dict1, start, finish):
        s=start
        visited_nodes = [] 
        queue = [(s,[s])]
        
        while len(queue)!=0:
            node, edge = queue.pop(0)
            visited_nodes.append(node)
           
            for i in dict1[node]:
                if (i not in visited_nodes) and (i != finish):
                    queue.append((i, list(edge + [i])))
                    visited_nodes.append(i)
                    #print(queue)
                elif (i == finish):
                    #print(list(edge + [finish]))
                    return len(list(edge + [finish]))-1
                else:
                    continue

file = open("file 2.txt")
#print(file.read())

for i in range(2):
    total_lines = int(file.readline().strip())


nodes_dictionary = {}
for i in range(int(total_lines)):
    values=file.readline().strip().split(" ")
    key, node = int(values[0]), int(values[1])
    if key in nodes_dictionary.keys():
        nodes_dictionary[key].append(node)
    else:
        nodes_dictionary[key] = [node] 
#print(nodes_dictionary)


participant = ["Lina", "Nora", "Lara"]
for i in range(len(participant)):
    participant[i]= int(file.readline())

NoraMoves = BFS(nodes_dictionary, participant[1], participant[0])
LaraMoves = BFS(nodes_dictionary, participant[2], participant[0])
#print(NoraMoves, LaraMoves)

if (LaraMoves < NoraMoves):
    print("Lara")
else:
    print("Nora")
```

### Task 03

Now, in this level, the participant is taken to “HighestLand” and is given a big challenge. He/she  discovers that there are a total of k competitors now (including him/her). All of their target is to  kill the warrior Lina in position ‘x’. But now, if there’s a connection from u to v, participants can  go from u to v but not vice versa. 
For n different fixed positions and m connections(u->v) among them, k different positions will  be given along with position ‘x’ and your task will be to find the participant who reaches ‘x’ first.  This time, you may just print the minimum number of moves the participant needed to reach  ‘x’. 
You are thinking of applying the BFS algorithm, right? Go on! But the main challenge is yet to come, you may apply the algorithm only once. So the complexity of your algorithm should be no less  than O(n+m). 




```python
file = open("file 3.txt")
#print(file.read())
total_pos_number = int(file.readline().strip())
total_connections = int(file.readline().strip())
#print(total_pos_number)
#print(total_connections)

node_list = []

for i in range(int(total_connections)):
    values=file.readline().strip().split(" ")
    temp_list = list([int(values[-1])] + [int(values[0])] + [int(1)])
    node_list.append(temp_list)

#print(node_list)

pos_Lina = int(file.readline().strip())
#print(pos_Lina)

participants_list = []
total_participants = int(file.readline().strip())

for i in range(total_participants):
    participants_list.append(int(file.readline().strip()))
#print(participants_list)

def ShortestMove(node_list, total_pos_number, participants_list, pos_Lina):
    distance = []
    for i in range(total_pos_number):
        if i!=pos_Lina:
            distance.append(float('inf'))
        else:
            distance.append(0)   
    #print(distance)
    
    for i in range(total_pos_number-1):
        for i1, i2, i3 in node_list:                           
                if distance[i1] is not float('inf'):                    
                    if distance[i2]> distance[i1] + i3:
                        distance[i2] = distance[i1] + i3

                else:
                    continue
      
    return list(distance)

move= ShortestMove(node_list, total_pos_number, participants_list, pos_Lina)
#print(move)

minimum_moves = float('inf')
#print(participants_list)
l = len(participants_list)
for i in range(l):    
    if (move[participants_list[i]] < minimum_moves):
        #print(move[participants[i]])
        minimum_moves = move[participants_list[i]]


print(minimum_moves)
```
