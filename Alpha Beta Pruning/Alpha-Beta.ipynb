```python
import random
import math
alpha = -math.inf 
beta = math.inf
pos = 0
depth = 0  
max_depth = int(input("Enter the Depth size: "))
branching_number = int(input("Enter the number of the nodes per branch: "))
total_leaf_node = branching_number**max_depth
#print(total_leaf_node)

number1 = str(input("Enter the minimum and maximum value for the range of the nodes in space separated way: "))
#print(number1)
range_list1 = number1.split(" ")
range_list2 = []
for i in range_list1:
    range_list2.append(int(i))
    
#print(range_list2)
max_range= max(range_list2)
min_range=min(range_list2)
#print("max_range", max_range)

leaf_value_list = []
for i in range(total_leaf_node):
    n = random.randint(min_range,max_range)
    leaf_value_list.append(n)
    
#print(leaf_value_list)  



def alphaBetaPruning(max_depth, depth, pos, branching_number, leaf_value_list, alpha, beta, maximizing):
    if depth == max_depth:
        return leaf_value_list[pos]
    if maximizing:
        maxEval = -math.inf
        for i in range(0, branching_number):
            value = alphaBetaPruning(max_depth, depth+1, pos*branching_number+i,branching_number,leaf_value_list, alpha, beta, False)
            maxEval = max(maxEval, value)
            alpha = max(alpha, maxEval)
            if alpha>=beta:
                break
        return maxEval
    
    else:
        minEval = +math.inf
        for i in range(0, max_depth-1):
            value = alphaBetaPruning(max_depth, depth+1, pos*branching_number+i, branching_number, leaf_value_list, alpha, beta, True)
            minEval = min(minEval, value)
            beta = max(beta, minEval)
            if alpha>=beta:
                break
        return minEval
        
print("Depth: ", 2*max_depth)
print("Branch: ", branching_number)
print("Terminal States:", len(leaf_value_list))
print("Maximum amount: ",alphaBetaPruning(max_depth, depth, pos, branching_number, leaf_value_list, alpha, beta, True))    
```
