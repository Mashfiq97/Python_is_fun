## OpenCV

```python
import numpy as np
import matplotlib.pyplot as plt

# %matplotlib inline makes your plot outputs appear and be stored 
# within the notebook. 
%matplotlib inline
```

### Import OpenCV

```python
import cv2
```

### Get the image with the imread 

```python
img = cv2.imread("Rock_Art.jpg")
```

### Always check the type

```python
type(img)
```

```python
img.shape
```

### Let's see the image file with the imshow

```python
plt.imshow(img)
```

### Matplotlib reads RGB
### OpenCV reads BGR
### We will convert OpenCV to the channels of the Photo

```python
img_fix = cv2.cvtColor( img, cv2.COLOR_BGR2RGB )
plt.imshow(img_fix)
```

### Scale it to gray and get shape

```python
img_gray = cv2.imread("Rock_Art.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# color dimension is not present
img_gray.shape
```

```python
# show the picture
plt.imshow(img_gray, cmap="gray")
```

### In OpenCV first Value is the width

```python
img_new = cv2.resize(img_fix, (700,1000))
plt.imshow(img_new)
```

### Resize with ratio

```python
width_ratio = 2
height_ratio = 2
img2 = cv2.resize(img_fix, (0,0), img_fix, width_ratio, height_ratio)
```

```python
plt.imshow(img2)
```

```python
img2.shape # doubling the shape
```

### Flip on Horizontal Axis

```python
img3 = cv2.flip(img_fix, 0)
plt.imshow(img3)
```

### Flip on Vertical Axis

```python
img3 = cv2.flip(img_fix, 1)
plt.imshow(img3)
```

# Draw Shapes on Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# %matplotlib inline makes your plot outputs appear and be stored 
# within the notebook. 
%matplotlib inline
```

### Create a Black Image

```python
black_img = np.zeros(shape=(512,512,3), 
                     dtype=np.int16)
```

```python
black_img.shape
```

```python
plt.imshow(black_img)
```

### Draw a Circle
#### The centre is the first number on X axis and seond on Y-axis

```python
cv2.circle(img=black_img, center=(400,100), radius=50, color=(255,0,0), thickness=8)
```

```python
plt.imshow(black_img)
```

### Filled the Circle

```python
cv2.circle(img=black_img, 
           center=(400,200), 
           radius=50, 
           color=(0,255,0), 
           thickness= -1)
plt.imshow(black_img)
```

### Draw a Rectangle

```python

# first number for x and second for y axis
# two points, one for up and 2nd for diagonally down
cv2.rectangle(black_img,
             pt1=(200,200),
             pt2=(300,300),
             color=(0,255,0),
             thickness=5)
plt.imshow(black_img)
```

### Write a Text

```python
import cv2
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(black_img,
           text="Mashfiq",
           org=(110,400),
           fontFace=font,
           fontScale=3,
           color=(255,255,0),
           thickness=3,
           lineType=cv2.LINE_AA)
plt.imshow(black_img)
```
